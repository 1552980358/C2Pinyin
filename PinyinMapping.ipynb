```kotlin
import java.io.File
import kotlin.io.encoding.Base64

fun generateBin(rootDir: File, base64s: Array<String>) {
    base64s.forEachIndexed { i: Int, str: String ->
        File(rootDir, "${i.inc()}.bin").let { file ->
            if (!file.exists()) {
                file.writeBytes(Base64.decode(str))
            }
        }
    }
}
```
